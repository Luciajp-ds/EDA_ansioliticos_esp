**1. INTRODUCCIÓN**

En los últimos años, la salud mental constituye uno de los principales desafíos sanitarios en Europa y, especialmente, en España. Concretamente, los trastornos de ansiedad se han consolidado, en las últimas décadas, como una de los principales problemas psicológicos, llegando a afectar a un 10% de la población española. Este aumento de malestar emocional coincide con una tendencia al alza en el uso de psicofármacos, especialmente de ansiolíticos y benzodiacepinas. Su uso se ha ido generalizando en diversos segmentos poblacionales y se ha convertido en una de las principales herramientas terapéuticas (Fontanet et al., 2024).

En este proyecto se aborda esta problemática desde una perspectiva analítica y exploratoria, utilizando datos oficiales para examinar la evolución del consumo de benzodiacepinas en España y los factores que pueden explicar su tendencia creciente (Claveras et al., 2022).




**2. TEMA ESCOGIDO**

Las benzodiacepinas constituyen uno de los grupos farmacológicos más prescritos en España para el tratamiento de la ansiedad, el insomnio y otros trastornos emocionales. En este contexto, España se sitúa entre los países europeos con mayor consumo de ansiolíticos desde hace más de veinte años. Concretamente, el consumo de benzodiacepinas ha pasado de 50,88% en 2012 a 58,09% en 2022, mientras que ya en 2005 España consumía un 25% más que la media europea, llegando incluso a duplicarla en 2015. Cabe señalar que las benzodiacepinas más utilizadas son lorazepam, alprazolam y diazepam (Calaveras et al., 2022; Fontanet et al., 2024; Bengoetxea et al., 2025; AEMP, 2021; AEMP, 2015; García et al., 2020).

Por otro lado, el consumo tiene patrones demográficos totalmente definidos: es notablemente mayor en mujeres (aproximadamente 2:1 con respecto a los hombres), e incrementa con la edad. No obstante, en los últimos años se está asistiendo a un incremento destacable del consumo de estas sustancias por parte de los adolescentes debido al aumento del malestar emocional (que coincide con un auge en los diagnosticos de ansiedad y depresión). 

Además, existen desigualdades socioeconómicas y territoriales claras: el uso es más elevado en municipios pequeños, zonas rurales y poblaciones con menor nivel de renta, lo que sugiere una relación entre vulnerabilidad social y mayor prescripción (Bengoetxea, et al., 2025; AEMP, 2021; Claveras et al., 2022).

En conclusión, la temática del presente EDA la constituye el estudio del consumo de benzodiacepinas en España, examinando:
- Diferencias por sexo, edad y nivel socioeconómico
- Su evolución a lo largo del tiempo (no se si lo haremos? /estos dos puntos se pueden modificar)

**3. JUSTIFICACIÓN**

El estudio del consumo de benzodiacepinas es relevante debido a que el mismo constituye un problema social y sanitario de creciente relevancia. A pesar de ser medicamentos eficaces, su uso prolongado conlleva riesgos como dependencia. En este contexto, España se sitúa entre los países europeos con mayor consumo de ansiolíticos, tendencia que se ha intensificado en la última década (Fontanet et al., 2024).

Los datos muestran que más del 30% de las mujeres y aproximadamente el 18% de los hombres mayores de 40 años recibieron al menos un envase de ansiolíticos en 2017. El consumo es superior en mujeres, incrementa con la edad y se vincula con factores socioeconómicos, siendo más elevado en zonas con menor renta y municipios poco poblados. En adolescentes, el aumento del malestar emocional (acentuado tras la pandemia) también ha favorecido un mayor uso de ansiolíticos  (Calaveras et al., 2022; Fontanet et al., 2024; Bengoetxea et al., 2025; AEMP, 2021; AEMP, 2015; García et al., 2020).

Entender estos patrones es imprescindible para poder orientar políticas sanitarias, así como diseñar estrategias preventivas. Por lo anterior, se justifica la elaboración de un análisis que nos permita estudiar la evolución, así como la distribución de su consumo de España a partir de diversas fuentes oficiales.




**4. OBJETIVO**

Por todo lo anteriormente mencionao, el objetivo principal de este EDA es analizar  la evolución del consumo de benzodiacepinas en España en los últimos años, identificando patrones por edad, sexo y nivel socioeconómico.

Para ello, se analizará dicho consumo, detectando patrones geográficos, tendencias temporales y outliers. Esto nos permitirá presentar un resumen estadístico y gráfico respecto el comportamiento relativo al uso de las mencionadas sustancias.

**5. HIPÓTESIS**

***Pongo varias hipótesis para que escojamos las que más nos gustan**

H1. El consumo de ansiolíticos es mayor en territorios con mayor densidad poblacional
- La concentración de población puede influir en la prescripción y dispensación de benzodiacepinas (mayor acceso a servicios de salud, mayor presión laboral y social).

H2. Las mujeres consumen más ansiolíticos que los hombres en toda slas franjas de edad
- Se baas en al evidencia de estudios y datos de aemps: las mujeres consumen casi el doble mas q hombres

H3. En los últimos años ha aumentado el consumo de ansiolíticos entre los adolescentes
- Hay estudios que muestran que como ha aumentado el malestar emocional + el diagnostico de ansiedad entre jovenes (sobre todo después de pandemia) = mayor consumo ansiolíticos

H4. Las personas con menor nivel de renta consumen más ansiolíticos
- influencia de factores socioeconómicos en el consumo + zonas con menor renta.

H5. El consumo de benzodiacepinas ha aumentado significativamente en los últimos años
- endencia general observada en los informes oficiales de la AEMPS (incremento en toda la población adulta)
